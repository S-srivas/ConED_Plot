```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from pylab import bar, show, plot

df=pd.read_csv('data/cned trial.csv', parse_dates=['START DATE'], index_col='START DATE')
m1_t = pd.read_csv('data/cned trial.csv',usecols = ['COST'])
```

```python
def plot_df(df, x, y, title="", xlabel='START DATE', ylabel='Usage', m1_tlabel='Cost', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.bar(x, y, color='purple')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)

file1 = plot_df(df, x=df.index, y=df.USAGE, title='Style 3')
file2 = m1_t    
plt.subplot(311)
plt.plot(x, file1)
plt.subplot(312, sharex=file1)
plt.plot(file2, y)  
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2,
                         ncols=1,
                         sharex=True)

df1 = pd.DataFrame(
    data = np.random.rand(25, 1),
    index=pd.date_range('2015-05-05', periods=25),
    columns=['DF1']
)

df2 = pd.DataFrame(
    data = np.random.rand(25, 1),
    index=pd.date_range('2015-04-10', periods=25),
    columns=['DF2']
)

df3 = pd.DataFrame(
    data = np.random.rand(50, 1),
    index=pd.date_range('2015-03-20', periods=50),
    columns=['DF3']
)

df3 = df3.reindex(index=df3.index.union(df2.index).union(df1.index))

df1.plot(ax=axes[0], linewidth=2, color='b', linestyle='solid')
df2.plot(ax=axes[0], linewidth=2, color='b', linestyle='dashed')
df3.plot(ax=axes[1])

plt.show()
```
